Ahora, vamos a comprobar la correlación entre los atributos.

```python
import numpy as np
import seaborn as sns
from sklearn.datasets import make_classification
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

# Generar datos de ejemplo
dataX, dataY = [col for col in wind_ava.columns if col.endswith('13')] + ['energy'],['energy']
X = wind_ava[dataX]
y = wind_ava[dataY]
# Calcular la matriz de correlación utilizando SelectKBest y f_classif
selector = SelectKBest(score_func=f_classif, k='all')
selector.fit(X, y)
correlation_matrix = np.corrcoef(X.T)

plt.figure(figsize=(30,25))
# Crear el heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', annot_kws={"size": 16})

# Mostrar el heatmap
variable_names = dataX
plt.xticks(ticks=np.arange(len(variable_names))+0.5, labels=variable_names, rotation=90)
plt.yticks(ticks=np.arange(len(variable_names))+0.5, labels=variable_names, rotation=0)

plt.title('Matriz de Correlación')
plt.show()
```

Podemos ver que hay atributos con una correlación muy elevada, incluso con una correlación del 100%. Por ello vamos a establecer un limite de correlación de máximo 98%.

```python
# Encontrar las variables con alta correlación
high_corr_indices = np.where(np.abs(correlation_matrix) > 0.98)
high_corr_pairs = [(variable_names[i], variable_names[j]) for i, j in zip(high_corr_indices[0], high_corr_indices[1]) if i != j]

# Decidir qué variable eliminar en cada par de variables altamente correlacionadas
variables_to_drop = set()
variables_to_keep = set()
correlated_pairs = set()
for i, j in high_corr_pairs:
    if i not in variables_to_drop and j not in variables_to_drop:
        correlation_ij = correlation_matrix[variable_names.index(i), variable_names.index(j)]
        if np.abs(correlation_ij) > 0.98:
            correlated_pairs.add((i, j))
            # Calcular la media de las correlaciones de cada variable con todas las demás
            avg_corr_i = np.abs(correlation_matrix[variable_names.index(i), :]).mean()
            avg_corr_j = np.abs(correlation_matrix[variable_names.index(j), :]).mean()
            if avg_corr_i > avg_corr_j:
                variables_to_drop.add(j)
                variables_to_keep.add(i)
            else:
                variables_to_drop.add(i)
                variables_to_keep.add(j)

print("Parejas de variables correlacionadas:", correlated_pairs)
print("Variables a mantener:", variables_to_keep)
print("Variables a eliminar:", variables_to_drop)
```

Eliminamos las variables correlacionadas.En este caso, Chat GPT intentó pasar a lista a la tabla de datos y después realizar esta resta de listas: columns_to_keep = list(set(columns_to_keep) - variables_to_drop)Sin embargo, no era correcto ya que de esta manera printear la tabla con índices y se movían los atributos de sitio como por ejemplo datetime.

```python
columns_to_keep = ['datetime','energy'] + [col for col in wind_ava.columns if col.endswith('13')]



for element in variables_to_drop:
  columns_to_keep.remove(element)
  print("Eliminamos: ", element)

sotavento = wind_ava[columns_to_keep]
sotavento
```

Como conclusión, hemos observado que no hay valores faltantes ni constantes. Hemos eliminado las variables con una correlación mayor al 98%. Y hemos podido observar que hay un hueco sin datos entre 2008 y 2009 ya que a cada uno le faltan los meses que tiene el otro.



# 2. Evaluación Outer (Estimación de Rendimiento Futuro / Evaluación de Modelo):

Dado que estamos tratando con un problema de regresión, las métricas principales que podríamos utilizar para evaluar el rendimiento del modelo son el Error Cuadrático Medio (MSE), el Error Absoluto Medio (MAE) y el Coeficiente de Determinación (R^2). Sin embargo, nos centraremos en el MSE solamente.

Procedimiento:

División de Datos: Utilizaremos los datos de los años 2008 y 2009 como conjunto de prueba (test set) y el resto de los datos como conjunto de entrenamiento (train set). Usaremos TimeSeriesSplit para trabajar con datos de series temporales. Divide el conjunto de datos en varios conjuntos de entrenamiento y prueba asegurándose de que los datos en el conjunto de prueba se encuentren después de los datos en el conjunto de entrenamiento en términos de tiempo. Esto ayuda a simular un escenario más realista donde el modelo se entrena con datos históricos y se evalúa con datos futuros.
# Evaluación Inner (Comparación de Alternativas y Ajuste de Hiperparámetros):
Para comparar diferentes alternativas y ajustar hiperparámetros, utilizaremos la validación cruzada.

Procedimiento:
Validación Cruzada: Utilizaremos el conjunto de entrenamiento para ajustar el modelo y seleccionar los hiperparámetros óptimos. Finalmente, evaluaremos el modelo con los hiperparámetros seleccionados en el conjunto de prueba y registraremos el rendimiento del modelo. Repetiremos este proceso para todas las iteraciones de la validación cruzada externa.

Selección de Modelos y Hiperparámetros:
Probaremos diferentes modelos de regresión (KNN, Regresión Lineal, Regresión Ridge, Regresión Lasso, Árboles de Regresión y SVM) y ajustaremos los hiperparámetros utilizando la validación cruzada interna en cada iteración de la validación cruzada externa.

Evaluación del Rendimiento:
Utilizaremos la métrica de evaluación (MSE) en el conjunto de prueba para cada iteración de la validación cruzada externa. Luego, calcularemos la media de esta métrica para obtener una estimación del rendimiento del modelo.

# 3. Elección método de escalado con KNN
Utilizando el algoritmo KNN (K-Nearest Neighbors), primero implementaremos la validación cruzada anidada. Luego, evaluaremos el rendimiento del modelo KNN con diferentes métodos de escalado para comprobar cual es más apropiado.

Validación Cruzada Anidada:

Dividiremos los datos en conjuntos de entrenamiento y prueba, utilizando los años 2008 y 2009 como conjunto de prueba y el resto como conjunto de entrenamiento.
Implementaremos la validación cruzada anidada para comparar los modelos y ajustar los hiperparámetros.

Métodos de Escalado:

Probaremos varios métodos de escalado, como la estandarización (StandardScaler) y la normalización (MinMaxScaler), con el modelo KNN.
Utilizaremos la validación cruzada anidada para evaluar el rendimiento del modelo KNN con cada método de escalado.
Compararemos los resultados para determinar el método de escalado más apropiado.

```python
from sklearn.model_selection import cross_val_score, TimeSeriesSplit
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.pipeline import Pipeline


X = sotavento[columns_to_keep]
y = sotavento['energy']

# Definir los datos de entrenamiento y prueba
train_years = [2004,2005,2006,2007]
test_years = [2008, 2009]
X['datetime'] = pd.to_datetime(X['datetime'])

# Separar los datos en conjuntos de entrenamiento y prueba
X_train = X[X['datetime'].dt.year.isin(train_years)].drop(['datetime','energy'], axis=1)
y_train = y[X['datetime'].dt.year.isin(train_years)]
X_test = X[X['datetime'].dt.year.isin(test_years)].drop(['datetime','energy'], axis=1)
y_test = y[X['datetime'].dt.year.isin(test_years)]

# Evaluacion interna con 3-fold CV
inner = TimeSeriesSplit(n_splits=3)
inner_scores = {}



# Alternativa 1: KNN con StandardScaler
pipeline_std = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsRegressor())
])
scores_std = cross_val_score(pipeline_std,X_train,y_train,cv=inner,scoring='neg_root_mean_squared_error')
inner_scores['KNN with StandardScaler'] = -scores_std.mean()

# Alternativa 2: KNN con MinMaxScaler
pipeline_minmax = Pipeline([
    ('scaler', MinMaxScaler()),
    ('knn', KNeighborsRegressor())
])
scores_std = cross_val_score(pipeline_minmax,X_train,y_train,cv=inner,scoring='neg_root_mean_squared_error')
inner_scores['KNN with MinMaxScaler'] = -scores_std.mean()

# Alternativa 3: KNN con RobustScaler
pipeline_robust = Pipeline([
    ('scaler', RobustScaler()),
    ('knn', KNeighborsRegressor())
])
scores_std = cross_val_score(pipeline_robust,X_train,y_train,cv=inner,scoring='neg_root_mean_squared_error')
inner_scores['KNN with RobustScaler'] = -scores_std.mean()

# Mostrar resultados
for name, score in inner_scores.items():
  print(f"{name}: {score}")
```

Hemos elegido el StandardScaler ya que es el mas adecuado para nuestro problema al tener un error menor. Una vez elegido el scaler, hacemos la estimación de rendimiento futuro:

```python
from sklearn import metrics
from math import sqrt
pipeline_std.fit(X_train,y_train)
y_pred = pipeline_std.predict(X_test)
outer_score = sqrt(metrics.mean_squared_error(y_test,y_pred))
print(f"Estimación de rendimiento de KNN con StandardScaler: {outer_score}")
```

KNN sin HPO --> Es el mismo que hicimos para calcular el scaler.

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.svm import SVR
import time
import copy
soluciones={}
modelo={}
# Definimos el primer modelo a utilizar, KNN
knn = KNeighborsRegressor()

# Escalamos los datos con RobustScaler
pipeline = Pipeline(steps=[('scaler', StandardScaler()), ('model', knn)])

# Entrenamos el modelo
grid_search = GridSearchCV(pipeline, param_grid = {}, cv=inner, scoring='neg_mean_squared_error', n_jobs=-1, verbose=1)
start_time = time.time()
grid_search.fit(X_train, y_train)
training_time = (time.time() - start_time)*1000
print(f"RMSE inner: {np.sqrt(-grid_search.best_score_)}")
# Hacer predicciones con el mejor modelo obtenido con grid search
y_pred = grid_search.best_estimator_.predict(X_test)
rmse = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
modelo["Sin HPO"]=rmse

print(f"RMSE outer: {rmse}")
print(f"Training time: {training_time} ms")
```

KNN con HPO --> Observamos que conseguimos mejorar levemente el error con el uso de HPO

```python
# Definir el pipeline
pipeline = Pipeline(steps=[('scaler', StandardScaler()), ('model', KNeighborsRegressor())])

# Definir los parámetros para la búsqueda en rejilla
param_grid_knn = {
    'model__n_neighbors': [3, 5, 7, 9, 11],  # Número de vecinos a usar
    'model__weights': ['uniform', 'distance'],  # Peso de los vecinos
    'model__p': [1, 2, 3],  # Potencia del parámetro Minkowski
    'model__metric': ['euclidean', 'manhattan'], # Métrica de distancia
    'model__algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']  # Algoritmo a utilizar
}

# Realizar la búsqueda en rejilla
grid_search = GridSearchCV(pipeline, param_grid_knn, cv=inner, scoring='neg_mean_squared_error', n_jobs=-1, verbose=1)
start_time = time.time()
grid_search.fit(X_train, y_train)
training_time = (time.time() - start_time)*1000
print(f"RMSE inner: {np.sqrt(-grid_search.best_score_)}")
# Hacer predicciones con el mejor modelo obtenido con grid search
y_pred = grid_search.best_estimator_.predict(X_test)

# Imprimir los resultados
print(f"Mejores parámetros: {grid_search.best_params_}")
rmse = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print(f"RMSE outer: {rmse}")
print(f"Training time: {training_time} ms")
modelo["Con HPO"]=rmse
modelo["hiperparametros"]=grid_search.best_params_
soluciones["KNN"]=modelo.copy()
```
